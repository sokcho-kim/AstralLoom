```python
import requests
import pandas as pd
import time

def get_all_reviews(app_id):
    reviews = []
    cursor = "*"
    has_more = True
    
    while has_more:
        url = f"https://store.steampowered.com/appreviews/{app_id}"
        params = {
            "json": 1,
            "language": "all",          # ëª¨ë“  ì–¸ì–´
            "filter": "recent",             # ì „ì²´ ë¦¬ë·°
            "day_range": 5,              # ì „ì²´ ê¸°ê°„
            "cursor": cursor,
            "num_per_page": 100,         # ìµœëŒ€ 100ê°œ
            "purchase_type": "all"       # êµ¬ë§¤ ì—¬ë¶€ ìƒê´€ì—†ì´
        }
        response = requests.get(url, params=params)
        
        if response.status_code == 200:
            data = response.json()
            cursor = data['cursor']       # ë‹¤ìŒ í˜ì´ì§€ ì»¤ì„œ ê°±ì‹ 
            has_more = data.get('success', 0) == 1 and len(data['reviews']) > 0  # ê³„ì† ë¦¬ë·°ê°€ ìˆìœ¼ë©´ ë°˜ë³µ
            for review in data['reviews']:
                reviews.append({
                    "review": review['review'],
                    "voted_up": review['voted_up'],
                    "timestamp_created": review['timestamp_created'],
                    "playtime_forever": review['author']['playtime_forever'],
                })
            time.sleep(1)  # ìŠ¤íŒ€ ì°¨ë‹¨ ë°©ì§€ ìœ„í•´ ì•½ê°„ ì‰¬ì–´ì£¼ì
        else:
            print(f"Error: {response.status_code}")
            break
    
    return pd.DataFrame(reviews)
```
